```python
surahs = []
```

```python
pages = np.arange(1, 115)
for page in pages:
    page = requests.get('https://quranenc.com/en/browse/vietnamese_hassan/'+ str (page) + ' ')
    soup = bs4.BeautifulSoup(page.text, 'lxml')
    toggle_content = soup.find_all('div', {'class' : 'toggle-content'})
    ttc = toggle_content[0].find_all('span', {'class' : 'ttc'})
    
    for translation in range(len(ttc)):
        printtext = ttc[translation].text

        surahs.append(printtext)
```

```python
len(surahs)
```

```python
surahs
```

```python
ts = pd.DataFrame({'Quran Translation' : surahs})
```

```python
ts
```

```python
ts = ts.dropna(how='any')
ts = ts.reset_index(drop=True)
```

```python
ts = ts.replace(r':', ' ', regex=True)
ts = ts.replace(r';', ' ', regex=True)
ts = ts.replace(r',', ' ', regex=True)
ts = ts.replace(r'!', ' ', regex=True)
ts = ts.replace(r'\'', ' ', regex=True)
ts = ts.replace(r'-', ' ', regex=True)
ts = ts.replace(r'"', ' ', regex=True)
ts = ts.replace(r'\?', ' ', regex=True)
ts = ts.replace(r'“', ' ', regex=True)
ts = ts.replace(r'”', ' ', regex=True)
ts = ts.replace(r'\n', ' ', regex=True)
ts = ts.replace(r'\.', ' ', regex=True)
ts = ts.replace(r'\(', ' ', regex=True)
ts = ts.replace(r'\)', ' ', regex=True)
ts = ts.replace(r'\d+', ' ', regex=True)
```
