```python
surahs = []
```

```python
pages = np.arange(1, 115)
for page in pages:
    page = requests.get('https://quranyusufali.com/'+ str (page) + ' ')
    soup = bs4.BeautifulSoup(page.text, 'lxml')
    entry_content = soup.find_all('div', {'class' : 'entry-content'})
    english_text = entry_content[0].find_all('div', {'class' : 'english-text'})
    
    for translation in range(len(english_text)):
        printtext = english_text[translation].text
        
        surahs.append(printtext)
```

```python
len(surahs)
```

```python
surahs
```

```python
ts = pd.DataFrame({'Quran Translation' : surahs}
```

```python
ts
```

```python
ts = ts.dropna(how='any')
ts = ts.reset_index(drop=True)
```

```python
ts = ts.replace(r';', ' ', regex=True)
ts = ts.replace(r':', ' ', regex=True)
ts = ts.replace(r',', ' ', regex=True)
ts = ts.replace(r'!', ' ', regex=True)
ts = ts.replace(r'\?', ' ', regex=True)
ts = ts.replace(r'-', ' ', regex=True)
ts = ts.replace(r'–', ' ', regex=True)
ts = ts.replace(r'‘', ' ', regex=True)
ts = ts.replace(r'’', ' ', regex=True)
ts = ts.replace(r'“', ' ', regex=True)
ts = ts.replace(r'”', ' ', regex=True)
ts = ts.replace(r'`', ' ', regex=True)
ts = ts.replace(r'\.', ' ', regex=True)
ts = ts.replace(r'\(', ' ', regex=True)
ts = ts.replace(r'\)', ' ', regex=True)
ts = ts.replace(r'…', ' ', regex=True)
ts = ts.replace(r'\d+', ' ', regex=True)
ts = ts.replace(r'\🙂', ' ', regex=True)
ts = ts.replace(r'\Section', ' ', regex=True)

#ts = ts.replace(r'\(.*\)', '', regex=True)
```
