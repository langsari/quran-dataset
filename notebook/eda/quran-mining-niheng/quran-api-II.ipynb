```python
import requests
import pandas as pd
```

```python
apijuzs = 'http://api.quran.com:3000/api/v3/juzs'
apichapters = 'http://api.quran.com:3000/api/v3/chapters'
```

<h1>apijuzs</h1>

```python
juz = requests.get(apijuzs)
juz.text
```

```python
juzI = pd.io.json.loads(juz.text)
type(juzI)
juzI
```

```python
juzI['juzs']
```

```python
d_juz = pd.DataFrame(juzI['juzs'])
d_juz
```

<h1>apichapters</h1>

```python
chapter = requests.get(apichapters)
chapter.text
```

```python
chapterI = pd.io.json.loads(chapter.text)
type(chapterI)
chapterI
```

```python
chapterI['chapters']
```

```python
d_chapter = pd.DataFrame(chapterI['chapters'])
d_chapter
```
